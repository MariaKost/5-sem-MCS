```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Построим графики зависимости T(P), S(P), E(P) при числе потоков N = 100000 и максимальном размере чанка m = 1000.

```python
P = np.array([1, 2, 4, 8, 16])
T_omp = np.array([0.04688, 0.03906, 0.03258, 0.02944, 0.03025])
T_thr = np.array([0.04688, 0.04075, 0.03249, 0.02871, 0.02711]) 
Qsort_t = 0.04818
S = T_thr[0] / T_thr 
E = S / P
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.ylabel('T')
plt.title('T(P)')
plt.plot(P, T_omp, '.-', label='Merge sort: OMP')
plt.plot(P, T_thr, '.-', label='Merge sort: Pthread')
plt.axhline(Qsort_t, label='Qsort', color = 'red')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.ylabel('S')
plt.title('S(P)')
plt.plot(P, S, '.-')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.ylabel('E')
plt.title('E(P)')
plt.plot(P, E, '.-')
plt.grid()
plt.show()
```
